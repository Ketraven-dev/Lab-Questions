```python
import numpy as np
import pandas as pd

data = {
    "roll_no": np.arange(1, 11),
    "marks": np.array([67, 87, 87, 67, 98, 89, 56, 76, 78, 90]),
    "Gender": ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F', 'M', 'F'],
    "attendance": np.array([8, 2, 4, 7, 7, 7, 3, 9, 10, 5])
}

df = pd.DataFrame(data)

print(df)
```

```python
null_val = df['marks'].isnull().sum()
print(f"Total null values: {null_val}")
```

```python
avg = df['marks'].mean()
print(f"Average is: {avg}")
```

```python
gender_count = df['Gender'].value_counts()
print(gender_count)
```

Question 2

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (6, 4))
plt.bar(gender_count.index, gender_count.values, color = ["blue", "pink"])
plt.xlabel("Gender")
plt.ylabel("No. of Students")
plt.title("Gender Distribution of Students")
plt.legend(["students"])
plt.show()
```

```python
plt.figure(figsize = (6, 5))
plt.hist(df['marks'], bins = 5, color = "yellow", edgecolor = "black")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.title("Histogram of students")
plt.legend(["students"])
plt.show()
```
