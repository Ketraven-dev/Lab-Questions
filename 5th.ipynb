```python
def dfs(st, goal, graph, path=None, vis=None):
    if path is None:
        path = []
    if vis is None:
        vis = set()

    path.append(st)
    vis.add(st)

    if st == goal:
        return path

    for nbr in graph.get(st, []):
        if nbr not in vis:
            res = dfs(nbr, goal, graph, path.copy(), vis)
            if res:
                return res

    return None
```

```python
def bfs(st, goal, graph, queue=None, vis=None):
    if queue is None:
        queue = [[st]]
    if vis is None:
        vis = {st}

    if not queue:
        return None

    path = queue.pop(0)
    node = path[-1]

    if node == goal:
        return path

    for nbr in graph.get(node, []):
        if nbr not in vis:
            vis.add(nbr)
            queue.append(path + [nbr])

    return bfs(st, goal, graph, queue, vis)

```

```python
st = 'A'
goal1 = 'C'
goal2 = 'D'
goal3 = 'F'

graph1 = {'A' : ['B', 'E', 'D'],
        'B' : ['A', 'E', 'C'],
        'C' : ['B', 'E', 'D'], 
        'D' : ['A', 'E', 'C'],
        'E' : ['A', 'B', 'C', 'D']}

graph2 = {'A': ['B', 'C'],
          'B': ['D'],
          'C': ['D'],
          'D': []}

graph3 = {'A': ['B', 'C'],
          'B': ['D', 'E'],
          'C': ['F'],
          'D': [],
          'E': ['F'],
          'F': []}
```

```python
print(bfs(st, goal1, graph1))
print("--------------------------------------")
print(dfs(st, goal1, graph1))
```

```python
print(bfs(st, goal2, graph2))
print("--------------------------------------")
print(dfs(st, goal2, graph2))
```

```python
print(bfs(st, goal3, graph3))
print("--------------------------------------")
print(dfs(st, goal3, graph3))
```
